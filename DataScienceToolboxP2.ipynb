### Introduction to Iterators

- Iterables 
  - Lists, Strings, Dictionaries, File connections are special objects called **iterables**
  - An object that has an associated `iter()` method
  - Applying `iter()` to an iterable creates an **iterator**
  - An iterable can be unpacked by the `iter()` function or traditional methods like `for` loops etc.
- Iterator
  - Object that has an associated `next()` method that produces next value with `next()`

```python
# How to create an iterator from an iterable
word = 'Da'
it = iter(word)

print(next(it))

print(next(it))
```

```python
print(next(it))
```

```python
# Print all values of an iterator at once
word = 'Data'
it = iter(word)
print(*it)

# Note that * is called the Splash operator
```

```python
# Once the iterator finishes unpacking all values, you can't do it again!
print(*it)
```

***An iterable is an object that can return an iterator, while an iterator is an object that keeps state and produces the next value when you call next()***

```python
# Create a list of strings: flash
flash = ['jay garrick', 'barry allen', 'wally west', 'bart allen']

# Print each list item in flash using a for loop
for names in flash:
    print(names)


# Create an iterator for flash: superspeed
superspeed = iter(flash)

# Print each item from the iterator
print(next(superspeed))
print(next(superspeed))
print(next(superspeed))
print(next(superspeed))
```

Recall that range() doesn't actually create the list; instead, it creates a range object with an iterator that produces the values until it reaches the limit (in the example, until the value 4). If range() created the actual list, calling it with a value of 1010010100 may not work, especially since a number as big as that may go over a regular computer's memory. The value 1010010100 is actually what's called a Googol which is a 1 followed by a hundred 0s. That's a huge number!

```python
# calling range() with 1010010100 won't actually pre-create the list.

# Create an iterator for range(10 ** 100): googol
googol = iter(range(10 ** 100))

# Print the first 5 values from googol
print(next(googol))
print(next(googol))
print(next(googol))
print(next(googol))
print(next(googol))
print(next(googol))
print(next(googol))
print(next(googol))
print(next(googol))
print(next(googol))
```

### enumerate() and zip()

- enumerate adds a counter to any iterator
- zip allows us to stitch together any number of iterables

```python
flash = ['jay garrick', 'barry allen', 'wally west', 'bart allen']

x = enumerate(flash)
x
```

```python
# x is an enumerate class object
# the values and the counter of an enumerate object can be returned as a list of tuples using the list() function
list(x) 
```

```python
# Start enumerate with a non-zero index
for x,y in enumerate(flash, start=10):
    print(x,y)
```

```python
random = ["blah1", "blah2", "blah3", "blah4"]
z = zip(flash, random)
```

```python
list(z)
```

```python
random2 = random + ["blah5"]
```

```python
# zip will only concatenate corresponding items so blah5 is not stitched
list(zip(flash, random2))
```

```python
for z1,z2 in zip(flash, random):
    print(z1,z2)
```

```python
# Create a list of tuples: mutant_data
mutants = ['charles xavier','bobby drake','kurt wagner','max eisenhardt','kitty pride']
aliases = ['prof x', 'iceman', 'nightcrawler', 'magneto', 'shadowcat']
powers = ['telepathy', 'thermokinesis', 'teleportation', 'magnetokinesis', 'intangibility']
mutant_data = list(zip(mutants, aliases, powers))

# Print the list of tuples
print(mutant_data)

# Create a zip object using the three lists: mutant_zip
mutant_zip = zip(mutants, aliases, powers)

# Print the zip object
print(mutant_zip)

# Unpack the zip object and print the tuple values
for value1, value2, value3 in zip(mutants, aliases, powers):
    print(value1, value2, value3)
```

*Remember that enumerate() and zip() are iterator objects and so once you are done unpacking them, further calls to unpack them will print nothing. For example....*

```python
z1 = zip(mutants, aliases)
```

```python
print(*z1)
```

```python
print(*z1)
```

```python
q1 = enumerate(mutants)
print(*q1)
```

```python
print(*q1)
```

There is no unzip function for doing the reverse of what zip() does. We can, however, reverse what has been zipped together by using zip() with a little help from \*! \* unpacks an iterable such as a list or a tuple into positional arguments in a function call.

```python
# 'Unzip' the tuples in z1 by unpacking with * and zip(): result1, result2
z1 = zip(mutants, aliases)
result1, result2 = zip(*z1)

# Check if unpacked tuples are equivalent to original tuples
print(result1)
print(result2)
```

```python
# Print a 5x5 matrix by the legacy for-loop way
import numpy as np
a = np.zeros((5,5), dtype=int)
for x in list(range(0,5)):
    for y in list(range(0,5)):
        a[x][y] = y
a
```

### Conditionals in Comprehensions

```python
# Conditions on the iterable
[x**2 for x in range(10) if x%2 == 0]
```

```python
# Conditions on the output expression
print([x**2 if x%2 == 0 else 0 for x in range(10)])

# Note that for this to work out, you have to use else!
print([x**2 if x%2 == 0 for x in range(10)])
```

### Introduction to generator expressions


- Generators are exactly like List Comprehensions except it does not store the list in memory.
- () instead of []
- Returns a Generator object
- But it can be iterated over

```python
result = (num for num in range(10))
for x in result:
    print(x)
```

```python
# We can also pass the Generator object to list() to create a list()
result = (num for num in range(10))
list(result)
```

```python
# Remember the context of enumerate and zip where the iterator would be exhausted of its values? Same applies
# for Generators. For example....
list(result)
```

*Generators come in handy when we are working with large datasets but we don't to load all of that onto memory like a List Comprehension would do!*

### Generator Functions

- You can create a generator function with the keyword **yield** in place **return** 
- Under the hood it will return a Generator Object (which is iterable)

```python
# Create generator object: result
result = (num for num in range(31))

# Print the first 5 values
print(next(result))
print(next(result))
print(next(result))
print(next(result))
print(next(result))
print("~~~~~~~~~~")

# Print the rest of the values
for value in result:
    print(value)
```
