### Python Lists

- Collection of values of similar type
- Can contain any type (ie. ints, floats, strings, booleans, lists, dictionaries etc.)
- Can contain different types within the list
- As opposed to int, bool etc., a list is a ***compound data type*** ie. you can group values together:
  - ***`a = 'is'`***
  - ***`b = 'nice'`***
  - ***`my_list = ['my', 'list', a, b]`***

```python
a = [1, 2, 3]
type(a)
```

```python
b = [1, 'two', [1,2,3], True] # List containing different types
print(b)
```

```python
c = [[1,2,3], [4,5,6]]
print(type(c))
```

Note the difference between the ***`type`*** of ***`list`*** and ***`list of list`***

### Subsetting and Slicing Lists

```python
a = [1, 2, 3, 4, 5]
a
```

```python
# First element
a[0]
```

```python
# Last Eelement
a[4]
```

```python
# 3rd element onwards
a[2:]
```

```python
# Up till the 3rd element excluding that
a[:2]
```

```python
# 2nd and 3rd element
a[1:3]
```

```python
# Last element
a[-1]
```

```python
# Last 3 elements
a[-3:]
```

```python
# 3rd last and 2nd elements
a[-3:-1]
```

```python
# Print everything
a[:]
```

Things become tricky when you copy lists. 

if 
    
    x = ['a', 'b', 'c']
    
then you are basically storing the list in your memory and the address/reference to that list in x. And so x does not actually contain the list but a reference to the list. Now if you do

    y = x
    
y will also reference that particular list in memory and so any changes to y will reflect in x!

```python
from IPython.display import Image
Image('reference.png', width=600, height=600)
```

If you want to create a new list y that points to a different address in the memory but with the same values, you can do 

- ***`y = list(x)`***

or

- ***`y = x[:]`***

And so now changes in y won't be reflected in x!
