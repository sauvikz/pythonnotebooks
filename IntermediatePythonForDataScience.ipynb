## Histograms

- Histograms are good for checking distributions of datasets

```python
# Given life_exp from above, plot Histogram (without specifying number of bins; by default it's 10)
plt.hist(life_exp)
plt.show()
```

The number of bins is pretty important. Too few bins will oversimplify reality and won't show you the details. Too many bins will overcomplicate reality and won't show the bigger picture.

```python
# Build a Histogram with 5 bins, clean it up and then build one with 20 bins
plt.hist(life_exp, bins=5)
plt.show()
plt.clf()

plt.hist(life_exp, bins=20)
plt.show()
plt.clf()
```

```python
# Draw the above 2 Histograms in the same plot
plt.hist(life_exp, bins=5, color="green", alpha=0.5)
plt.hist(life_exp, bins=20, color="blue", alpha=0.5)
plt.show()
```

## Pandas!

- High level data manipulation tool
- Wes McKinney
- Built on Numpy
- DataFrame


- df.index = ['', '', ....] to set row indices
- pd.read_csv('filepath', index_col = 0) to set the 0th column as row indices

Square brackets
- Column access: brics[["country", "capital"]]
- Row access: only through slicing brics[1:4]

`loc` (label-based; more versatile and in sync with 2D numpy arrays!)
 - Row access: brics.loc[["RU", "IN", "CH"]]
 - Column access: brics.loc[:, ["country", "capital"]]
 - Row & Column access: brics.loc[["RU", "IN", "CH"], ["country", "capital"]]
 
`iloc` is exactly the same as `loc` except that you use numbered indices and not labeled indices for both `rows` and `columns`
 - Row access: brics.loc[[1,2,3]]
 - Column access: brics.loc[:, [0,1]]
 - Row & Column access: brics.loc[[1,2,3], [0,1]]
 
 
You might wonder if you can also combine label-based selection the loc way and index-based selection the iloc way. You can! It's done with ix, but we won't go into that here.

If `cars` is a DataFrame,

`cars['cars_per_cap']`
`cars[['cars_per_cap']]`

The single bracket version gives a Pandas Series, the double bracket version gives a Pandas DataFrame.

##### String comparison operators:
- "Carl" < "Chris" wil give you True because in lexical order, "Carl" comes before "Chris"
- Comparison can only be done for objects of the same type for example 3 < 4 or 3 < 4.1 (even though is int and double)
- But True == 1 and False == 0 (It just works that way)
- True >= False

##### Boolean Operators
- and
- or
- not

`not` has a higher priority than `and` & `or` Also remember that `and` is logical and `&` is bitwise!

This does not work for Numpy arrays although. For example, if x is an array of numbers then `x > 5 and x < 20` will not work. Hence we need to use:
- logical_and()
- logical_or()
- logical_not()

Hence the above would be `logical_and(x > 5, x < 20)`

The same logic can be applied for subsetting Pandas DataFrames; for example:

- `df[np.logical_or(df['col'] > 10, df['col'] < 20)]`

```python
import pandas as pd

cars_per_cap = [809, 731, 588, 18, 200, 70, 45]
country = ["United States", "Australia", "Japan", "India", "Russia", "Morocco", "Egypt"]
drives_right = [True, False, False, False, True, True, True]
ind = ["US", "AUS", "JAP", "IN", "RU", "MOR", "EG"]

cars_dict = {
    "cars_per_cap": cars_per_cap,
    "country": country,
    "drives_right": drives_right
}

cars = pd.DataFrame(cars_dict)
cars.index = ind
print(cars)
```

```python
# Subset the above DataFrame based on whether cars_per_cap > 500
print(cars[cars['cars_per_cap'] > 500])

# or
print(cars[cars.loc[:,"cars_per_cap"] > 500])

# or
print(cars[cars.iloc[:, 0] > 500])

# Remember if you are subsetting based on a single column, you have to select iloc and loc in such a way that a 
# Series and not a DataFrame is selected
```

```python
#  Select the observations that have a cars_per_cap between 10 and 80
import numpy as np
print(cars[np.logical_and(cars['cars_per_cap'] > 10, cars['cars_per_cap'] < 80)])

# or 
print(cars[np.logical_and(cars.loc[:, 'cars_per_cap'] > 10, cars.loc[:, 'cars_per_cap'] < 80)])

# or 
print(cars[np.logical_and(cars.iloc[:, 0] > 10, cars.iloc[:, 0] < 80)])
```
